```python
import numpy as np
import pandas as pd

import os
```

```python
DATA_PATH = '../data/'
RANDOM_SEED = 1337
```

```python
os.listdir(DATA_PATH)
```

```python
dataframes = {}
for file in os.listdir(DATA_PATH):
    name = file.split(".")[0]
    print(name)
    dataframes[name] = pd.read_csv(os.path.join(DATA_PATH, file))
    display(dataframes[name])
```

```python
dataframes["merc"] = dataframes["merc"].append(dataframes["cclass"])
dataframes["ford"] = dataframes["ford"].append(dataframes["focus"])
del dataframes["focus"], dataframes["cclass"]
dataframes["ford"]
```

```python
dataframes["hyundi"].rename(columns={"tax(£)": "tax"}, inplace=True)
dataframes["hyundi"]
```

```python
columns = [
    "make", "model", "year", "price", "transmission", "mileage", "fuelType",
    "tax", "mpg", "engineSize"
]
df = pd.DataFrame(columns=columns)

for name in dataframes:
    file_df = dataframes[name]
    file_df["make"] = name
            
    df = df.append(file_df)
df
```
